**WEATHER FORECASTING FOR CHECKING TORNADO,CYCLONE,HURRICANCE .**
**WE USED ARIMA AND SARIMA APPROACH FOR FORECASTING,ARIMA DEAL WITH FORECASTING WITH HISTORICAL ASPECT AND SARIMA DEAL WITH FORECASTING WITH TAKING ACCOUNT THE SEASONAL COMPONENT.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import ipywidgets as widgets
from IPython.display import display

# Load and preprocess your historical weather data
# Replace 'your_data.csv' with the actual path to your data
df = pd.read_csv('/content/weather.csv', parse_dates=['Date'], index_col='Date')
# df.dropna(inplace=True)

# Convert relevant columns to numeric data types
numeric_columns = ['WindHighMPH', 'SeaLevelPressureLowInches']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Subset the data for actual data (January to July) and forecast period (August)
actual_data = df['2017-07-01':'2017-07-31']
forecast_data = df['2017-08-01':'2017-08-15']

# Define ARIMA parameters (p, d, q)
arima_p = 1
arima_d = 0  # No differencing
arima_q = 1

# Define SARIMA parameters (p, d, q) and seasonal parameters (P, D, Q, S)
sarima_p = 1
sarima_d = 0  # No differencing
sarima_q = 1
sarima_P = 1
sarima_D = 1
sarima_Q = 1
sarima_S = 7  # Weekly seasonality

# Fit the ARIMA model using actual data
arima_model = ARIMA(actual_data['WindHighMPH'], order=(arima_p, arima_d, arima_q))
arima_result = arima_model.fit()

# Fit the SARIMA model using actual data
sarima_model = SARIMAX(actual_data['WindHighMPH'], order=(sarima_p, sarima_d, sarima_q),
                       seasonal_order=(sarima_P, sarima_D, sarima_Q, sarima_S))
sarima_result = sarima_model.fit()

# Create interactive widgets
forecast_period_slider = widgets.IntSlider(value=len(forecast_data), min=1, max=len(forecast_data), step=1, description='Forecast Period')
button = widgets.Button(description="Update Plots")

def update_plots(button_click):
    forecast_period = forecast_period_slider.value

    # Forecast for the specified period using ARIMA
    arima_forecast = arima_result.get_forecast(steps=forecast_period)
    arima_forecasted_wind_speed = arima_forecast.predicted_mean

    # Forecast for the specified period using SARIMA
    sarima_forecast = sarima_result.get_forecast(steps=forecast_period)
    sarima_forecasted_wind_speed = sarima_forecast.predicted_mean

    # Plot ARIMA forecast
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(actual_data.index, actual_data['WindHighMPH'], label='Actual Data (Jan-Jul)')
    plt.plot(forecast_data.index, arima_forecasted_wind_speed, color='black', label='ARIMA Forecast (Aug)')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed (mph)')
    plt.title('ARIMA Wind Speed Forecast')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    # Plot SARIMA forecast
    plt.subplot(1, 2, 2)
    plt.plot(actual_data.index, actual_data['WindHighMPH'], label='Actual Data (Jan-Jul)')
    plt.plot(forecast_data.index, sarima_forecasted_wind_speed, color='red', label='SARIMA Forecast (Aug)')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed (mph)')
    plt.title('SARIMA Wind Speed Forecast')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

    # Print forecasted wind speeds for ARIMA and SARIMA
    for date, arima_wind_speed, sarima_wind_speed in zip(forecast_data.index, arima_forecasted_wind_speed, sarima_forecasted_wind_speed):
        print(f"{date}: ARIMA Forecasted Wind Speed = {arima_wind_speed:.1f} mph, SARIMA Forecasted Wind Speed = {sarima_wind_speed:.1f} mph")

# Connect the button to the function
button.on_click(update_plots)

# Display widgets
display(forecast_period_slider, button)
```

**CHECKING ALL FEATURES OF PARTICULAR CANDIDATE ID REGARDING CANDIDATE CAN AFFORD INSURANCE OR NOT,PROFILE SCORE ,HOUSE VALUE AND CAR AGE.**

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display


# Load your dataset
#cf = pd.read_csv("/content/car_insurance_claim.csv")

# Extract relevant columns for analysis
selected_columns = ['ID', 'INCOME', 'MSTATUS', 'KIDSDRIV']
data = cf[selected_columns].copy()

# Convert 'INCOME' column to numeric after removing '$' and commas
data['INCOME'] = pd.to_numeric(data['INCOME'].str.replace('[$,]', ''), errors='coerce')

# Calculate expenses based on marital status and number of kids
average_expenses_per_person = 800  # Adjust this value as needed

def calculate_expenses(row):
    marital_status = row['MSTATUS']
    kids = row['KIDSDRIV']
    if marital_status == 'z_No':
        expenses = average_expenses_per_person * (1 + kids)
    else:
        expenses = average_expenses_per_person * (2 + kids)
    return expenses

data['Expenses'] = data.apply(calculate_expenses, axis=1)

# Calculate available income after deducting expenses from income
data['Available_Income'] = data['INCOME'] - data['Expenses']

# Extract relevant columns for profile score analysis
selected_columns_profile = ['ID', 'CLM_FREQ', 'CLM_AMT']
profile_data = cf[selected_columns_profile].copy()

# Convert 'CLM_AMT' column to numeric after removing '$' and commas
profile_data['CLM_AMT'] = pd.to_numeric(profile_data['CLM_AMT'].str.replace('[$,]', ''), errors='coerce')

# Calculate profile score based on claim frequency
def calculate_profile_score(row):
    claim_freq = row['CLM_FREQ']
    if claim_freq == 0:
        return 'Good'
    elif claim_freq == 1:
        return 'OK'
    elif claim_freq > 1:
        return 'Not OK'
    else:
        return 'Undefined'

profile_data['Profile_Score'] = profile_data.apply(calculate_profile_score, axis=1)

# Extract relevant columns for information display
selected_columns_info = ['ID', 'HOME_VAL', 'CAR_USE', 'CAR_AGE']
info_data = cf[selected_columns_info].copy()

# Convert 'HOME_VAL' column to numeric after removing '$' and commas
info_data['HOME_VAL'] = pd.to_numeric(info_data['HOME_VAL'].str.replace('[$,]', ''), errors='coerce')

# Define candidate ID input box for all features
id_input = widgets.IntText(
    description='Candidate ID:'
)
# Define labels for displaying results
candidate_affordability_label = widgets.Label(value='')
profile_score_label = widgets.Label(value='')
info_label = widgets.Label(value='')
# Define an input widget for insurance amount
insurance_input = widgets.FloatText(
    description='Insurance Amount:'
)


def check_all_features(_):
    candidate_id = id_input.value

    # Check Affordability
    candidate_row_afford = data[data['ID'] == candidate_id]
    if not candidate_row_afford.empty:
        insurance_amount = insurance_input.value
        can_afford = candidate_row_afford.iloc[0]['Available_Income'] >= insurance_amount
        candidate_affordability_label.value = f'Candidate {candidate_id} can afford: {can_afford}'
    else:
        candidate_affordability_label.value = f'Candidate ID {candidate_id} not found for affordability'

    # Check Profile Score
    candidate_row_profile = profile_data[profile_data['ID'] == candidate_id]
    if not candidate_row_profile.empty:
        profile_score = candidate_row_profile.iloc[0]['Profile_Score']
        profile_score_label.value = f'Profile Score for Candidate {candidate_id}: {profile_score}'
    else:
        profile_score_label.value = f'Candidate ID {candidate_id} not found for profile score'

    # Show Info
    candidate_row_info = info_data[info_data['ID'] == candidate_id]
    if not candidate_row_info.empty:
        house_info = ''
        car_info = ''
        if not pd.isnull(candidate_row_info.iloc[0]['HOME_VAL']):
            house_info = f'House Value: ${candidate_row_info.iloc[0]["HOME_VAL"]:.2f}'
        car_use = candidate_row_info.iloc[0]['CAR_USE']
        car_age = candidate_row_info.iloc[0]['CAR_AGE']
        if car_use == 'Private' and not pd.isnull(car_age):
            car_info = f'Car Age: {int(car_age)} years'
        info_label.value = f'{house_info}\n{car_info}'
    else:
        info_label.value = f'Candidate ID {candidate_id} not found for information display'

# Check button to trigger all feature checks
check_button_all = widgets.Button(description='Check All Features')
check_button_all.on_click(check_all_features)

# Display input box and check button for all features
display(id_input)
display(check_button_all)
display(candidate_affordability_label)
display(profile_score_label)
display(info_label)
```

**RISK SCORING AND NEW PREMIUM PRICING AFTER COMPARING THE RISK SCORES.
RISK SCORING PARAMETERS ARE :- WEATHER RISK, AFFORDABILITY ,PROFILE SCORE(HISTORICAL CLAIM FREQUENCY),ASSET STRENGTH(HOUSE VALUE),CAR AGE.**

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import ipywidgets as widgets
from IPython.display import display
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load your df and cf data frames and preprocess as needed
df = pd.read_csv("/content/weather.csv")  # Replace with your data loading code
cf = pd.read_csv("/content/car_insurance_claim.csv")  # Replace with your data loading code
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['WindHighMPH'] = pd.to_numeric(df['WindHighMPH'], errors='coerce')

# Function to forecast wind speeds using ARIMA
def arima_forecast(data, forecast_period):
    model = ARIMA(data, order=(1, 0, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=forecast_period)
    return forecast

# Function to forecast wind speeds using SARIMA
def sarima_forecast(data, forecast_period):
    model = SARIMAX(data, order=(1, 0, 1), seasonal_order=(1, 1, 1, 7))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=forecast_period)
    return forecast

# Calculate weather risk level based on forecasts
def calculate_weather_risk(arima_forecast_values, sarima_forecast_values):
    arima_risk_score = max(arima_forecast_values)
    sarima_risk_score = max(sarima_forecast_values)
    final_risk_score = max(arima_risk_score, sarima_risk_score)

    if final_risk_score >= 14:
        risk_level = 'RED'
    elif final_risk_score >= 10:
        risk_level = 'YELLOW'
    else:
        risk_level = 'GREEN'

    return risk_level

# Function to calculate expenses based on marital status and number of kids
def calculate_expenses(row):
    marital_status = row['MSTATUS']
    kids = row['KIDSDRIV']
    average_expenses_per_person = 800  # Adjust this value as needed

    if marital_status == 'z_No':
        expenses = average_expenses_per_person * (1 + kids)
    else:
        expenses = average_expenses_per_person * (2 + kids)

    # Calculate average expenses for kids based on their ages
    if kids > 0:
        # Calculate average age of kids
        average_age_of_kids = 30 + (2 * (kids - 1))
        kid_expenses = average_expenses_per_person * (1 + average_age_of_kids / 100)
        expenses += kid_expenses

    return expenses

# Function to check affordability
def check_affordability(candidate_id, insurance_amount):
    candidate_row_afford = cf[cf['ID'] == candidate_id]
    if not candidate_row_afford.empty:
       # Convert 'INCOME' to numeric after removing '$' and commas
        candidate_income = pd.to_numeric(candidate_row_afford.iloc[0]['INCOME'].replace('[$,]', ''), errors='coerce')

        # Calculate expenses based on marital status and number of kids
        candidate_expenses = calculate_expenses(candidate_row_afford.iloc[0])

        # Calculate available income after deducting expenses from income
        available_income = candidate_income - candidate_expenses

        if available_income >= insurance_amount:
            afford_status = 'GREEN'
        elif available_income >= 0.9 * insurance_amount:
            afford_status = 'YELLOW'
        else:
            afford_status = 'RED'
    else:
        afford_status = 'Candidate ID not found for affordability'

    return afford_status

    # Calculate profile score function
def calculate_profile_score(candidate_id):
    candidate_row_cf = cf[cf['ID'] == candidate_id]
    if not candidate_row_cf.empty:
        claim_frequency = candidate_row_cf.iloc[0]['CLM_FREQ']
        if claim_frequency == 0:
            profile_score = 'GREEN'
        elif claim_frequency <= 2:
            profile_score = 'YELLOW'
        else:
            profile_score = 'RED'
    else:
        profile_score = None  # Candidate ID not found
    return profile_score

# Calculate financial asset strength function
def calculate_financial_asset_strength(candidate_id):
    candidate_row_cf = cf[cf['ID'] == candidate_id]
    if not candidate_row_cf.empty:
        house_value = pd.to_numeric(candidate_row_cf.iloc[0]['HOME_VAL'].replace('[$,]', ''), errors='coerce')
        if house_value >= 150000:
            asset_strength = 'GREEN'
        elif house_value >= 100000:
            asset_strength = 'YELLOW'
        else:
            asset_strength = 'RED'
    else:
        asset_strength = None  # Candidate ID not found
    return asset_strength

# Calculate car age risk function
#def calculate_car_age_risk(car_age):
def calculate_car_age_risk(candidate_id):
    candidate_row_cf = cf[cf['ID'] == candidate_id]
    if not candidate_row_cf.empty:
       car_age= candidate_row_cf.iloc[0]['CAR_AGE']
       if car_age <= 5:
          car_age_risk = 'GREEN'
       elif car_age <= 10:
            car_age_risk = 'YELLOW'
       else:
         car_age_risk = 'RED'
    else:
       car_age_risk = None  # Candidate ID not found
    return car_age_risk

# Create input widgets
candidate_id_input = widgets.IntText(description='Candidate ID:')
forecast_date_input = widgets.DatePicker(description='Forecast Date:')
insurance_amount_input = widgets.FloatText(description='Insurance Amount:')
calculate_button = widgets.Button(description='Calculate')

# Create output widgets
weather_risk_level_output = widgets.Label(value='')
affordability_status_output = widgets.Label(value='')
profile_score_output = widgets.Label(value='')
asset_strength_output = widgets.Label(value='')
car_age_risk_output = widgets.Label(value='')
# Create output widget for premium calculation
premium_output = widgets.Label(value='')

# Function to calculate premium based on risk factors
def calculate_premium(weather_risk, affordability, profile_score, asset_strength, car_age, insurance_amount):
    risk_scores = {
        'GREEN': 4,
        'YELLOW': 2,
        'RED': 1
    }

    weather_score = risk_scores.get(weather_risk, 0)
    affordability_score = risk_scores.get(affordability, 0)
    profile_score_score = risk_scores.get(profile_score, 0)
    asset_strength_score = risk_scores.get(asset_strength, 0)
    car_age_score = risk_scores.get(car_age, 0)

    total_score = weather_score + affordability_score + profile_score_score + asset_strength_score + car_age_score

    if total_score == 12:
        premium = insurance_amount
    elif 5 <= total_score < 12:
        premium = insurance_amount * 1.45  # Increase by 45%
    else:
        premium = insurance_amount * 2.20  # Increase by 120%

    return premium


# Calculate function
def calculate_button_click(change):
    candidate_id = candidate_id_input.value
    forecast_date = pd.Timestamp(forecast_date_input.value)  # Convert to Timestamp
    insurance_amount = insurance_amount_input.value

    forecast_period = (forecast_date - pd.Timestamp('2017-08-01')).days + 1
    arima_forecast_values = arima_forecast(df['WindHighMPH'], forecast_period)
    sarima_forecast_values = sarima_forecast(df['WindHighMPH'], forecast_period)
    weather_risk = calculate_weather_risk(arima_forecast_values, sarima_forecast_values)
    affordability = check_affordability(candidate_id, insurance_amount)
    profile_score = calculate_profile_score(candidate_id)
    asset_strength = calculate_financial_asset_strength(candidate_id)
    car_age = calculate_car_age_risk(candidate_id)
    weather_risk_level_output.value = f"Weather Risk Level: {calculate_weather_risk(arima_forecast_values, sarima_forecast_values)}"
    affordability_status_output.value = f"Affordability: {check_affordability(candidate_id, insurance_amount)}"
    profile_score_output.value = f"Profile Score: {calculate_profile_score(candidate_id)}"
    asset_strength_output.value = f"Financial Asset Strength: {calculate_financial_asset_strength(candidate_id)}"
    #car_age_risk_output.value = f"Car Age Risk: {calculate_car_age_risk(candidate_id)}"
    car_age_risk_output.value = f"Car Age Risk: {calculate_car_age_risk(candidate_id)}"
     # Calculate premium
    premium = calculate_premium(weather_risk, affordability, profile_score, asset_strength, car_age, insurance_amount)
    premium_output.value = f"Premium of Insurance: {premium}"

calculate_button.on_click(calculate_button_click)

# Display widgets
widgets_box = widgets.VBox([
    candidate_id_input,
    forecast_date_input,
    insurance_amount_input,
    calculate_button,
    weather_risk_level_output,
    affordability_status_output,
    profile_score_output,
    asset_strength_output,
    car_age_risk_output,
    premium_output

])
display(widgets_box)
```
